```python
import numpy as np
X=np.array([-2,-2,1,2])
Y=np.array([-2,-1,2,1])
```

```python
import matplotlib.pyplot as plt
Xa = X**2
Ya = X*Y

```

```python
U=np.array([-2,0,0,2])
V=np.array([2,2,-1,-1])
Xb = U**2
Yb = U*V
```

```python
P=np.array([-2,-2,1,2])
Q=np.array([-2,-1,2,1])
R=np.array([-2,0,0,2])
S=np.array([2,2,-1,-1])
plt.scatter(P,Q,color="orange")
plt.scatter(R,S,color="green")
plt.title("input space")
plt.show()
```

```python
plt.scatter(Xa,Ya, color="red")
plt.scatter(Xb,Yb,color="blue")
plt.title("feature space")
plt.show()










```
