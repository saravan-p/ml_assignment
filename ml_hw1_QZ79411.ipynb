```python
def g(z1, z2, a, b):
    return (a-z1)**2 + b*(z2-z1**2)**2
```

```python
# partial derivative w.r.t z1
def g1(z1, z2, a, b): 
    return 4*b*(z1**3) - 4*b*z1*z2 + 2*z1 - 2*a

# partial derivative w.r.t z2
def g2(z1, z2, a, b): 
    return 2*b*(z2 - z1**2)
```

```python
def gradient_descent(learning_rate, z1_init, z2_init, err_threshold=1e-4):
    z1_prev, z2_prev = z1_init, z2_init
    a, b = 1, 100
    err = 1e5 # initialize error as a very large value

    iterations = []
    i = 0
    func_values = []
    errors = []
    
    while err > err_threshold:  # change this stop criterion as per convenience
        # gradient descent step for z1
        z1 = z1_prev - (learning_rate * g1(z1_prev, z2_prev, a, b))
        
        # gradient descent step for z2
        z2 = z2_prev - (learning_rate * g2(z1_prev, z2_prev, a, b))
        
        # calculate the function value
        func_values.append(g(z1, z2, a, b))

        # count the number of iterations
        i = i+1
        iterations.append(i)

        # calculate the distance between previous and current (z1,z2) pairs
        err = ((z1-z1_prev)**2 + (z2-z2_prev)**2)**0.5  
        errors.append(err)
        
        print(z1, z2, err)
        
        z1_prev, z2_prev = z1, z2
    
    return z1, z2, func_values, iterations,errors
```

```python
z1_opt, z2_opt, g_values, iters, errors = gradient_descent(learning_rate=1e-4, z1_init=2, z2_init=1)
```

```python
g(z1_opt, z2_opt, a=1, b=100)
```
